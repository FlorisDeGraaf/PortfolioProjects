Use BeautifulSoup to scrape data from a wikipedia article about Premier League records and Statistics. Populate the variable 'soup' with this page.

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Premier_League_records_and_statistics'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

Find the correct table from the Wikipedia article, in this case the 6th table. Then use the 'th' (table header) to find the headers. Populate them in a variable and print to see if it is done correctly.

```python
soup.find('table')
```

```python
table = soup.find_all('table')[6]
```

```python
footy_titles = table.find_all('th')
```

```python
footy_titles
```

Clean up the variable using Strip.

```python
footy_table_titles = [title.text.strip() for title in footy_titles]

print(footy_table_titles)
```

Import pandas and create a table using the headers from 'footy_table_titles'

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = footy_table_titles)

df
```

Now it's time to find the actual data we need to use. In the code we can see within the table, 'tr' is used for every table row and 'td' is used within the 'tr' to host the individual cell data. We are going to use find_all on every 'tr' within the 'table' variable (not 'soup'), then we are going to use a for loop to find each 'td' within each 'tr', put them in a list(individual_row_data) and immediately put them into the dataframe. We do this because every time the for loop goes through the table rows, the variable 'individual_row_data' is overwritten by the next row.

We also start the for loop from position [1], the second position, because position [0] is an empty list. This wouldn't work as the dataframe expects a certain number of fields per list, all other rows do own a that number of fields.

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
